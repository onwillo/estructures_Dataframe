```python
import requests
from bs4 import BeautifulSoup

# URL de la portada de Wikipedia en español
url = "https://es.wikipedia.org/wiki/Wikipedia:Portada"

# Realizar la solicitud HTTP GET a la página web
response = requests.get(url)

# Crear un objeto BeautifulSoup para analizar el contenido HTML
soup = BeautifulSoup(response.text, "html.parser")

# Encontrar el contenedor principal de la portada
portada = soup.find("div", {"id": "mw-content-text"})

# Encontrar los elementos relevantes dentro de la portada
elementos = portada.find_all(["h2", "h3", "h4", "p"])

# Mostrar los elementos encontrados
print("Elementos encontrados:")
for elemento in elementos:
    print(elemento.get_text())

```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup

# Ruta al controlador de ChromeDriver
ruta_controlador = 'C:/Users/Willo/Desktop/chromedriver.exe'

# Configuración de opciones de Chrome
opciones = Options()
opciones.add_argument('--headless')  # Para ejecutar en modo sin cabeza (sin ventana del navegador visible)

# Configurar el servicio del controlador de ChromeDriver
servicio = Service(ruta_controlador)

# Inicializar el controlador de ChromeDriver
driver = webdriver.Chrome(service=servicio, options=opciones)

# URL del sitio a analizar
url = 'https://www.bolsasymercados.es/bme-exchange/es/Indices/Ficha/Ibex-35-ES0SI0000005'

# Abrir la página en el navegador
driver.get(url)

# Obtener el contenido HTML de la página
contenido_html = driver.page_source

# Cerrar el navegador
driver.quit()

# Crear un objeto BeautifulSoup para analizar el contenido HTML
soup = BeautifulSoup(contenido_html, 'html.parser')

# Imprimir la estructura HTML
print(soup.prettify())

```
