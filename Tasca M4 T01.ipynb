```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer archivo CSV
df = pd.read_csv('barri.csv')

# Agrupar por distrito y contar los barrios
barrios_por_distrito = df.groupby('Nom_Districte')['Nom_Barri'].nunique()

# Calcular porcentaje de cada distrito
porcentaje_por_distrito = (barrios_por_distrito / barrios_por_distrito.sum()) * 100

# Crear gráfico de torta
plt.pie(porcentaje_por_distrito.values, labels=porcentaje_por_distrito.index, autopct='%1.1f%%')
plt.title('Participación porcentual de cada distrito por numeros de barrios')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar datos en una tabla
df = pd.read_excel('nuevobarri.xls')

# Agrupar datos por nombre y trimestre y sumar los precios
grouped_data = df.groupby(['Nom_Districte', 'Trimestre'])['Preu'].sum()

# Reorganizar datos en una tabla con nombre y trimestre como índices y precios como valores
grupos = grouped_data.unstack()

# Generar gráfico apilado
grupos.plot(kind='bar', stacked=True)

# Agregar título y etiquetas a los ejes
plt.title('Precios totalizados por trimestre y distritos')
plt.xlabel('Trimestre')
plt.ylabel('Precios')

# Mostrar gráfico
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# practica con matriz de datos y dos variables en x
datos = {
    'distrito': ['distrito A', 'distrito B', 'distrito C', 'distrito A', 'distrito C'],
    'barrio': ['Barrio 1', 'Barrio 2', 'Barrio 3', 'Barrio 4', 'Barrio 5'],
    'presupuesto': [1000, 1500, 800, 1200, 900]
}

df = pd.DataFrame(datos)

# Agrupar y graficar
agrupado = df.groupby('distrito')['presupuesto'].mean().reset_index()


plt.bar(agrupado['distrito'], agrupado['presupuesto'])


plt.xlabel('Distrito')
plt.ylabel('Presupuesto promedio')
plt.title('Presupuesto promedio por Distrito BCN')


plt.show()
```
